```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
data1.drop(columns=["NOMBRE"], inplace=True)
data2.drop(columns=["NOMBRE"], inplace=True)
```

```python
X_train = data1.loc[:, data1.columns != "PRECIO"]
y_train = data1.loc[:, "PRECIO"]

X_test = data2.loc[:, data2.columns != "PRECIO"]
y_test = data2.loc[:, "PRECIO"]
```

```python
clf = DecisionTreeRegressor()
clf.fit(X_train,y_train)
y_train_pred = clf.predict(X_train)
print("Train accuracy:",mean_squared_error(y_train, y_train_pred))
y_test_pred = clf.predict(X_test)
print("Test  accuracy:",mean_squared_error(y_test, y_test_pred))
```

```python
from sklearn.model_selection import train_test_split
X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, train_size=0.70, random_state=42)
```

```python
from sklearn.metrics import accuracy_score

def calcular_accuracy_train_val(clf, X_train, y_train, X_val, y_val):
    clf.fit(X_train,y_train)
    y_train_pred = clf.predict(X_train)
    print("Entrenamiento accuracy:",accuracy_score(y_train, y_train_pred))
    y_val_pred = clf.predict(X_val)
    print("Validacion accuracy:",accuracy_score(y_val, y_val_pred))
    return clf
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error

clf.fit(X_train,y_train)
y_train_pred = clf.predict(X_train)
print("R2")
print("Entrenamiento accuracy:",r2_score(y_train, y_train_pred))
y_test_pred = clf.predict(X_test)
print("Prueba accuracy:",r2_score(y_test, y_test_pred))
print("MSE")
print("Entrenamiento accuracy:",mean_squared_error(y_train, y_train_pred))
y_test_pred = clf.predict(X_test)
print("Prueba accuracy:",mean_squared_error(y_test, y_test_pred))
print("MSLE")
print("Entrenamiento accuracy:",mean_squared_log_error(y_train, y_train_pred))
y_test_pred = clf.predict(X_test)
print("Prueba accuracy:",mean_squared_log_error(y_test, y_test_pred))
```
